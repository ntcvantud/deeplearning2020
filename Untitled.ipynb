```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
def diffx2(x):
    return 2*x

def x2(x):
    return x**2

def GDx2(x0,learning_rate):
    x = [x0]
    cost = [x2(x0)]
    for i in range(1000):
        learned_x = x[-1] - learning_rate*diffx2(x[-1])
        x.append(learned_x)
        cost.append(x2(learned_x))
    return (x,cost)


fig = plt.figure(figsize=(20,5))

graph = GDx2(-4,0.99)
plt.subplot(1,3,1)
plt.xlabel('too high')
plt.plot(graph[0],graph[1])

graph = GDx2(-4,0.1)
plt.subplot(1,3,2)
plt.xlabel('just right')
plt.plot(graph[0],graph[1])

graph = GDx2(-4,0.00001)
plt.subplot(1,3,3)
plt.xlabel('too low')
plt.plot(graph[0],graph[1])
;
```

```
w = np.array([1.,1.]).reshape(-1,1)
numOfIteration = 100
cost = np.zeros((numOfIteration,1))
learning_rate = 0.000001
X = [w.tolist()[0]]
Y = [w.tolist()[1]]
for i in range(1, numOfIteration):
    r = np.dot(x, w) - y
    cost[i] = 0.5*np.sum(r*r)
    w[0] -= learning_rate*np.sum(r)
    w[1] -= learning_rate*np.sum(np.multiply(r, x[:,1].reshape(-1,1)))
    X.append(w.tolist()[0])
    Y.append(w.tolist()[1])
```

Cu 4 Use Gradient Descent for data_square.csv

```
data = pd.read_csv('/content/drive/My Drive/Copy of data_square.csv').values
N = data.shape[0]
x = data[:, 0].reshape(-1, 1)
y = data[:, 1].reshape(-1, 1)
plt.scatter(x, y)
plt.xlabel('mt vung')
plt.ylabel('gi')
```

```
x = data[:, 0].reshape(-1, 1)
y = data[:, 1].reshape(-1, 1)
x = np.stack([x, np.square(x)], axis=-1)
x = x.reshape(x.shape[0], x.shape[-1])
x = np.hstack((np.ones((N, 1)), x))
```

```
w_itr = np.array([1.,1.,1.]).reshape(-1,1)
w_model = w_itr
numOfIteration = 100              

cost_itr = np.zeros((numOfIteration,1))
cost_model = 1e100
learning_rate = 0.00000001
while(1):
    for i in range(1, numOfIteration):
        r = np.dot(x, w_itr) - y
        cost_itr[i] = 0.5*np.sum(r*r)
        w_itr[0] -= learning_rate*np.sum(r)/30
        w_itr[1] -= learning_rate*np.sum(np.multiply(r, x[:,1].reshape(-1,1)))/30
        w_itr[2] -= (learning_rate**2)*np.sum(np.multiply(r, x[:,2].reshape(-1,1)))/30
    if cost_itr[-1][0] < 19000:
        w_model = w_itr
        break
    elif cost_itr[-1][0] < cost_model:
        learning_rate *= 1.005
        w_model = w_itr
        cost_model = cost_itr[-1][0]
    elif cost_itr[-1][0] > cost_model:
        learning_rate *= 0.85
        w_itr = w_model
```
